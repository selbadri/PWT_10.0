```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Load the Data
pwt = pd.read_csv('https://raw.githubusercontent.com/jivizcaino/PWT_10.0/main/pwt100.csv', encoding="latin-1")
pwt
```

```python
#Prepare the Dataset and Clean the Data
pwt_subset = pwt[['country','countrycode','year','cgdpo','pop','emp','avh','hc','cn','labsh','ctfp']]
pwt_subset = pwt_subset[pwt_subset['year'] > 2009]
pwt_subset = pwt_subset.set_index('country')
#pwt_subset = pwt_subset.dropna()
pwt_subset
```

```python
#Choosing the Year to Perform the Analysis
print(pwt_subset.groupby("year").count())
```

We will choose 2019: the latest year with the largets number of ebservations for our variables of interest.

```python
#Descriptive Statistics for the Variables in our Dataset
print(pwt_subset[pwt_subset['year'] == 2019].describe([0.05,0.10,0.25,0.50,0.75,0.90,0.95]).round(2))
```

```python
pwt_subset_2019 = pwt_subset[pwt_subset['year'] == 2019]
pwt_subset_2019
```

# Development Accounting

## GDP per worker

```python
pwt_subset_2019['GDP_pw'] = pwt_subset_2019['cgdpo']/pwt_subset_2019['emp']
pwt_subset_2019['GDP_pw'].describe([0.01,0.05,0.10,0.25,0.50,0.75,0.90,0.95,0.99]).round(2)
```

```python
pwt_subset_2019
```

```python
#pwt_subset_2019['GDP_pc'] = pwt_subset_2019['cgdpo']/pwt_subset_2019['pop']
pwt_subset_2019[['GDP_pc','GDP_pw','ctfp']].describe([0.01,0.05,0.10,0.25,0.50,0.75,0.90,0.95,0.99]).round(2)
```

```python
pwt_subset_2019_3vars = pwt_subset_2019[['GDP_pc','GDP_pw','ctfp']]
pwt_subset_2019_3vars.dropna()
```

```python
var_log_ypw = np.var(np.log(pwt_subset_2019_3vars['GDP_pw']))
var_log_ypw
```

```python
var_log_ypc = np.var(np.log(pwt_subset_2019_3vars['GDP_pc']))
```

```python
var_log_tfp = np.var(np.log(pwt_subset_2019_3vars['ctfp']))
```

```python
var_log_tfp/var_log_ypw
```

```python
pwt_subset_2019_7vars = pwt_subset_2019[['GDP_pc','GDP_pw','ctfp','cn','emp','labsh','hc']]
pwt_subset_2019_6vars = pwt_subset_2019[['GDP_pc','GDP_pw','ctfp','cn','emp','hc']]
pwt_subset_2019_6vars = pwt_subset_2019_6vars.dropna()
pwt_subset_2019_7vars = pwt_subset_2019_7vars.dropna()
```

```python
pwt_subset_2019_6vars['cn']
pwt_subset_2019_6vars['emp']
```

```python
pwt_subset_2019_6vars['k'] = (pwt_subset_2019_6vars['cn']/pwt_subset_2019_6vars['emp'])
pwt_subset_2019_6vars['y_hk_pw'] = (pwt_subset_2019_6vars['k']**(1/3))*((pwt_subset_2019_6vars['hc'])**(2/3))
pwt_subset_2019_6vars['y_hk_pw'].describe([0.01,0.05,0.10,0.25,0.50,0.75,0.90,0.95,0.99]).round(2)
```

```python
np.var(np.log(pwt_subset_2019_6vars['y_hk_pw']))
```

```python
np.var(np.log(pwt_subset_2019_6vars['ctfp']))
```

```python
np.var(np.log(pwt_subset_2019_6vars['GDP_pw']))
```
